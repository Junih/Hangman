# HANGMAN MINI GAME

```python
import random
import sys
from IPython.display import Markdown, display

def print_bold(string):
    display(Markdown(string))

def hangman_hard():
    print("    ________")
    print("    |    |")
    print("    |    ")
    print("    |    ")
    print("    |    ")
    print("    |    ")
    print("    |    ")
    print("    |    ")
    print(" ________" "\n")
    
def hangman_medium():
    print("    _____")
    print("    |    ")
    print("    |    ") 
    print("    |    ")
    print("    |    ")
    print("    |    ")
    print("    |    ")
    print("    |    ")
    print(" ________" "\n")
    
def hangman_easy():
    print("    ")
    print("    |    ")
    print("    |    ")
    print("    |    ")
    print("    |    ")
    print("    |    ")
    print("    |    ")
    print("    |    ")
    print(" ________" "\n")
    

def difficulty_set():
    easy_tab = ["easy", "E", "EASY", "EZ", "ez", "Easy"]
    medium_tab = ["MEDIUM", "M", "medium", "normal", "Medium"]
    hard_tab = ["HARD", "H", "hard", "Hard"]
    tries_easy = 13
    tries_medium = 8
    tries_hard = 4
    print("Easy - ", str(tries_easy), "lives")
    print("Medium - ", str(tries_medium), "lives")
    print("Hard - ", str(tries_hard), "lives")
    print_bold("**Choose the difficulty:** ")
    difficulty = input()
    print("")
    while True:
        if difficulty in easy_tab:
            print("You have chosen easiest difficulty and will receive", str(tries_easy), "tries." "\n")
            start_playing_easy()
        elif difficulty in medium_tab:
            print("You have chosen medium difficulty and will receive", str(tries_medium), "tries." "\n")
            start_playing_medium()
        elif difficulty in hard_tab:
            print("You have chosen hardest difficulty and will receive", str(tries_hard), "tries." "\n")
            start_playing_hard()
        else:
            print("Invalid characters! " "\n")
            difficulty_set()

def main_menu():
    quit_tab = ["quit", "QUIT", "Quit", "Q", "q"]
    mode_tab = ["Mode", "mode", "MODE", "M", "m"]
    welcome_string = print_bold('**Welcome to hangman!** ')
    print("Mode - To select difficulty")
    print("Quit - Exits a program" "\n")
    print_bold("**Which action do you want to perform:** ")
    action = input()
    print("")
    if action in quit_tab:
        sys.exit()
    elif action in mode_tab:
        difficulty_set()
    else:
        print("Invalid characters! " "\n")
        main_menu()

def get_word_hard():
    words = ["business", "intelligence", "success", "depression", "maintenance", "glimmer", "warchief"]
    return random.choice(words)        
        
def get_word_medium():
    words = ["pineapple", "developer", "programming", "engineering", "information", "technology", "amusement"]
    return random.choice(words)        
        
def get_word_easy():
    words = ["monkey", "apple", "banana", "game", "hate", "love", "Anger", "hangman"]
    return random.choice(words)

def retry():
    yes_tab = ["y", "yes", "Y", "YES", "YA"]
    no_tab = ["n", "no", "N", "NO", "NA"]
    print_bold("**Get back to main menu? Y/N** ")
    answer = input()
    if answer in yes_tab:
        main_menu()
    elif answer in no_tab:
        sys.exit()
    else:
        print("Invalid characters! " "\n")
        retry()

def start_playing_hard():
    alpha = "abcdefghijklmnopqrstuvwxyz"
    word = get_word_hard()
    tries = 4
    letters_guessed = []
    gameover = False
    print("Looking word contains", len(word), "letters, good luck!" "\n")
    print(len(word) * '_ ')
    print("")
    print("You have", str(tries), "tries to guess the word." "\n")
    hangman_hard()
    while gameover == False and tries > 0:
        guess = input("Enter the letter or complete word: ").lower()
        # OPTION 1 - USER INPUTS THE LETTER
        if len(guess) == 1:
            if guess in word:
                print_bold("**Correct!** ")
                letters_guessed.append(guess)
            elif guess not in word:
                print("Wrong. ")
                letters_guessed.append(guess)
                tries -= 1
                print("You have", str(tries), "left.")
            elif guess not in alpha:
                print("A-Z characters only! " "\n")
            elif guess in letters_guessed:
                print("You tried this letter before! " "\n")
            else:
                print("Try again" "\n")
        #OPTION 2 - USER INPUTS COMPLETE WORD
        elif len(guess) == len(word):
            if guess == word:
                print_bold("**Congratulations, you have guessed the word!** " "\n")
                gameover = True
            else:
                print("Wrong. ")
                tries -= 1
                print("You have", str(tries), "left.")
        #OPTION 3 - USER INPUTS WORD WITH LENGTH SMALLER OR BIGGER THAN LENGTH OF THE LOOKING WORD
        else:
            print("You must insert either a letter or word with same length as word you are trying to guess. " "\n")
            tries -= 1
            print("You have", str(tries), "left.")
    
        status = ''
        if gameover == False:
            for letter in word:
                if letter in letters_guessed: # if letters from word are in letters_guessed table, add them to "status" string
                    status += letter
                else: # if letters from word are not in letters_guessed table, display the letter as "_"
                    status += '_ '
            print("Looking word:", status)
            print("")
        if status == word:
            print_bold("**Congratulations, you have guessed the word!** " "\n")
            gameover = True
        elif tries == 0:
            print_bold("**RIP [*]"** "\n")
            gameover = True
            
        print("    ________")
        print("    |    |")
        print("    |    " + ("o" if tries < 4 else "")) 
        print("    |    " + ("/\\" if tries < 3 else ""))
        print("    |    " + ("|" if tries < 2 else ""))
        print("    |    " + ("/\\" if tries < 1 else ""))
        print("    |    ")
        print("    |    ")
        print(" ________" "\n")
        
    retry()
        

def start_playing_medium():
    alpha = "abcdefghijklmnopqrstuvwxyz"
    word = get_word_medium()
    tries = 8
    letters_guessed = []
    gameover = False
    hangman_medium()
    print("Looking word contains", len(word), "letters, good luck!" "\n")
    print(len(word) * '_ ')
    print("")
    print("You have", str(tries), "tries to guess the word." "\n")
    while gameover == False and tries > 0:
        guess = input("Enter the letter or complete word: ").lower()
        # OPTION 1 - USER INPUTS THE LETTER
        if len(guess) == 1:
            if guess in word:
                print_bold("**Correct!** ")
                letters_guessed.append(guess)
            elif guess not in word:
                print("Wrong. ")
                letters_guessed.append(guess)
                tries -= 1
                print("You have", str(tries), "left.")
            elif guess not in alpha:
                print("A-Z characters only! " "\n")
            elif guess in letters_guessed:
                print("You tried this letter before! " "\n")
            else:
                print("Try again" "\n")
        #OPTION 2 - USER INPUTS COMPLETE WORD
        elif len(guess) == len(word):
            if guess == word:
                print_bold("**Congratulations, you have guessed the word!** " "\n")
                gameover = True
            else:
                print("Wrong. ")
                tries -= 1
                print("You have", str(tries), "left.")
        #OPTION 3 - USER INPUTS WORD WITH LENGTH SMALLER OR BIGGER THAN LENGTH OF THE LOOKING WORD
        else:
            print("You must insert either a letter or word with same length as word you are trying to guess. " "\n")
            tries -= 1
            print("You have", str(tries), "left.")
    
        status = ''
        if gameover == False:
            for letter in word:
                if letter in letters_guessed: # if letters from word are in letters_guessed table, add them to "status" string
                    status += letter
                else: # if letters from word are not in letters_guessed table, display the letter as "_"
                    status += '_ '
            print("Looking word:", status)
            print("")
        if status == word:
            print_bold("**Congratulations, you have guessed the word!** " "\n")
            gameover = True
        elif tries == 0:
            print_bold("**RIP [*]**" "\n")
            gameover = True
            
        print("    _____"+("_" if tries < 8 else "")+("_" if tries < 7 else "")+ ("_" if tries < 6 else ""))
        print("    |    " +("|" if tries < 5 else ""))
        print("    |    " + ("o" if tries < 4 else "")) 
        print("    |    " + ("/\\" if tries < 3 else ""))
        print("    |    " + ("|" if tries < 2 else ""))
        print("    |    " + ("/\\" if tries < 1 else ""))
        print("    |    ")
        print("    |    ")
        print(" ________" "\n")
        
    retry()
        
def start_playing_easy():
    alpha = "abcdefghijklmnopqrstuvwxyz"
    word = get_word_easy()
    tries = 13
    letters_guessed = []
    gameover = False
    hangman_easy()
    print("Looking word contains", len(word), "letters, good luck!" "\n")
    print(len(word) * '_ ')
    print("")
    print("You have", str(tries), "tries to guess the word." "\n")
    while gameover == False and tries > 0:
        guess = input("Enter the letter or complete word: ").lower()
        # OPTION 1 - USER INPUTS THE LETTER
        if len(guess) == 1:
            if guess in word:
                print_bold("**Correct!** ")
                letters_guessed.append(guess)
            elif guess not in word:
                print("Wrong. ")
                letters_guessed.append(guess)
                tries -= 1
                print("You have", str(tries), "left.")
            elif guess not in alpha:
                print("A-Z characters only! " "\n")
            elif guess in letters_guessed:
                print("You tried this letter before! " "\n")
            else:
                print("Try again" "\n")
        #OPTION 2 - USER INPUTS COMPLETE WORD
        elif len(guess) == len(word):
            if guess == word:
                print_bold("**Congratulations, you have guessed the word!** " "\n")
                gameover = True
            else:
                print("Wrong. ")
                tries -= 1
                print("You have", str(tries), "left.")
        #OPTION 3 - USER INPUTS WORD WITH LENGTH SMALLER OR BIGGER THAN LENGTH OF THE LOOKING WORD
        else:
            print("You must insert either a letter or word with same length as word you are trying to guess. " "\n")
            tries -= 1
            print("You have", str(tries), "left.")
    
        status = ''
        if gameover == False:
            for letter in word:
                if letter in letters_guessed: # if letters from word are in letters_guessed table, add them to "status" string
                    status += letter
                else: # if letters from word are not in letters_guessed table, display the letter as "_"
                    status += '_ '
            print("Looking word:", status)
            print("")
        if status == word:
            print_bold("**Congratulations, you have guessed the word!** " "\n")
            gameover = True
        elif tries == 0:
            print_bold("**RIP [*]**" "\n")
            gameover = True
            
        print("    "+("_" if tries < 13 else "")+("_" if tries < 12 else "")+("_" if tries < 11 else "")+("_" if tries < 10 else "")+("_" if tries < 9 else "")+("_" if tries < 8 else "")+("_" if tries < 7 else "")+ ("_" if tries < 6 else ""))
        print("    |    " +("|" if tries < 5 else ""))
        print("    |    " + ("o" if tries < 4 else "")) 
        print("    |    " + ("/\\" if tries < 3 else ""))
        print("    |    " + ("|" if tries < 2 else ""))
        print("    |    " + ("/\\" if tries < 1 else ""))
        print("    |    ")
        print("    |    ")
        print(" ________" "\n")
            
    retry()
    

main_menu()
```
